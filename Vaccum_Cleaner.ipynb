```python
import random
import numpy as np

class Environment(object):
    def __init__(self):
        self.Ground = {'A':0, 'B':0, 'C':0, 'D':0, 'E':0}
        self.First = {'A':0, 'B':0, 'C':0, 'D':0, 'E':0}
        self.Second = {'A':0, 'B':0, 'C':0, 'D':0, 'E':0}
        # instantiate locations and conditions
        # 0 indicates  Clean and 1 indicates Dirty
        self.locationCondition = {'Ground':0, 'First':0, 'Second':0}


        # randomize conditions for floors
        self.locationCondition['Ground'] = random.randint(0, 1)
        self.locationCondition['First'] = random.randint(0, 1)
        self.locationCondition['Second'] = random.randint(0,1)

        if(self.locationCondition['Ground'] == 1):
          self.Random_Cleaning(self.Ground)
        if(self.locationCondition['First'] == 1):
          self.Random_Cleaning(self.First)
        if(self.locationCondition['Second'] == 1):
          self.Random_Cleaning(self.Second)
        if(self.locationCondition['First'] == 0 and self.locationCondition['Second'] == 0 and self.locationCondition['Ground'] == 0):
          print("All Floors are Clean")

    def Random_Cleaning(self,Floor):
      for i in Floor:
        Floor[i] = random.randint(0,1)
      print(Floor)

class SimpleReflexVacuumAgent(Environment):
    def __init__(self, Environment):
        print (Environment.locationCondition)
        # Instantiate performance measurement
        Score = 0
        # place vacuum at Ground Floor location
        vacuumLocation = 0
        count = 0
        battery = 100 # in percent
        storage = 1000 # in kilogram

        # If vacuum  at Ground Floor
        if vacuumLocation == 0:
          count = 0
          print("\n\t********* Vacuum Cleaner Cleaning Ground floor ************")

          if Environment.locationCondition['Ground'] == 0:
            print("\n\t%%%%%%%% Good Ground Floor is already Clean %%%%%%%%")

          else:
            for i in Environment.Ground:
              if Environment.Ground[i] == 1:
                print("Room "+str(i)+" is Dirty")


                #Suck and make Clean
                if(storage==0):
                  print("storage becomes 0  storage need to replace \n")
                  print("replacing replacing...... \n replaced now storaage bag is empty \n")
                  storage = 1000
                if(storage > 0):
                  storage = storage - 200
                  print("storage becomes full by 200 gram\n")

                if(battery==0):
                  print("battery becomes 0 percent battery need to charge \n")
                  print("kept on charging charging...... \n battery becomes 100 percent \n")
                  battery = 100
                if(battery > 0):
                  battery = battery - 20
                  print("battery deducted by 20 percent\n")
                  Environment.Ground[i] = 0
                  Score += 1
                  print ("Location A has been Cleaned. Moving Forward ---->")

              else:
                print("Room "+str(i)+" is Already Clean. Moving Forward ---->")


              count += 1
          vacuumLocation = 1

        # Now Vacuum Cleaner is at First Floor
        if vacuumLocation == 1:

          # Penalty for moving form ground floor to first
          Score -= 2

          count = 0
          # self.Visulization(Environment,1,0)

          print("\n\t********** Vacuum Cleaner Cleaning First floor *************")

          # If floor is clean already
          if Environment.locationCondition['First'] == 0:
            print("\n\t%%%%%%%% Good First Floor is already Clean %%%%%%%%")

          # If floor is dirty, then visit each room and clean it
          else:
            for i in Environment.First:
              if Environment.First[i] == 1:
                print("Room "+str(i)+" is Dirty")


                #Suck and make Clean
                if(storage==0):
                  print("storage becomes 0  storage need to replace \n")
                  print("replacing replacing...... \n replaced now storaage bag is empty \n")
                  storage = 1000
                if(storage > 0):
                  storage = storage - 200
                  print("storage becomes full by 200 gram\n")

                if(battery==0):
                  print("battery becomes 0 percent battery need to charge \n")
                  print("kept on charging charging...... \n battery becomes 100 percent\n")
                  battery = 100
                if(battery > 0):
                  battery = battery - 20
                  print("battery deducted by 20 percent\n")
                  Environment.Ground[i] = 0
                  Score += 1
                  print ("Location A has been Cleaned.Moving Forward ---->")


              else:
                print("Room "+str(i)+"is Already Clean. Moving Forward ---->")


              count += 1
          vacuumLocation = 2

        # Now Cleaning agent is at Second Floor
        if vacuumLocation == 2:
          # penalty for moving First to Second floor
          Score -= 2

          count = 0


          print("\n\t********** Vacuum Cleaner Cleaning Second floor *************")

          if Environment.locationCondition['Second'] == 0:
            print("\n\t%%%%%%%% Good Second Floor is already Clean %%%%%%%%")

          else:
            for i in Environment.Second:
              if Environment.Second[i] == 1:
                print("Room "+str(i)+" is Dirty")


                #Suck and make Clean
                if(storage==0):
                  print("storage becomes 0  storage need to replace \n")
                  print("replacing replacing...... \n replaced now storaage bag is empty \n")
                  storage = 1000
                if(storage > 0):
                  storage = storage - 200
                  print("storage becomes full by 200 gram\n")

                if(battery==0):
                  print("battery becomes 0 percent battery need to charge \n")
                  print("kept on charging charging...... \nbattery becomes 100 percent \n")
                  battery = 100
                if(battery > 0):
                  battery = battery - 20
                  print("battery deducted by 20 percent\n")
                  Environment.Ground[i] = 0
                  Score += 1
                  print ("Location "+str(i)+ " has been Cleaned. Moving Forward ---->")


              else:
                print("Room "+str(i)+" is Already Clean. Moving Forward ---->")


              count += 1

        print("\n\t^^^^^^^^^ All Rooms of all Floors are cleaned ^^^^^^^^^^")
        print("\t\tPerformance:- "+str(Score))
        print("\t\t last updated battery condition :- "+str(battery))
        print("\t\t last storage bag condition :- "+str(storage))



theEnvironment = Environment()
theVacuum = SimpleReflexVacuumAgent(theEnvironment)

```
